```python
import json
import pandas as pd
```

```python
with open('../data/raw/data.json', encoding='utf-8') as f:
    data = json.load(f)
```

```python
messages = data['messages']
len(messages)
```

```python
texts = [
    m for m in messages
    if 'text' in m and isinstance(m['text'], str) and m['text'].strip() != ''
]
len(texts)
```

```python
df_raw = pd.DataFrame(messages)
df_raw.head()
```

```python
df_raw.shape
```

```python
df_raw['type'].value_counts()
```

```python
df_raw['text'].apply(type).value_counts()
```

```python
df_raw[df_raw['text'].apply(lambda x: x == '')].shape[0]
```

```python
def extract_text_len(x):
    if isinstance(x, str):
        return len(x)
    if isinstance(x, list):
        return sum(len(i.get('text', '')) for i in x if isinstance(i, dict))
    return 0

df_raw['text_len'] = df_raw['text'].apply(extract_text_len)
df_raw['text_len'].describe()
```

```python
df_messages = df_raw[
    (df_raw['type'] == 'message') &
    (df_raw['text_len'] > 20)
]
df_messages.shape
```
